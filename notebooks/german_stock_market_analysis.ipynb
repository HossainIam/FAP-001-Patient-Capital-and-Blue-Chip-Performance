# FAP-001: German Stock Market Analysis

## Professional Financial Analytics Portfolio Project

This project analyzes selected DAX-listed companies using historical stock market data from Yahoo Finance.

The objective is to evaluate performance, risk characteristics, sector differences, and investment insights from a financial analyst's perspective.


## 1. Market Introduction

The German stock market is one of the most important equity markets in Europe. It includes large multinational companies from sectors such as technology, industrials, insurance, telecommunications, and automotive manufacturing.

This project focuses on selected major German companies that are part of the DAX index. The analysis will compare their historical stock performance, risk levels, return characteristics, and sector-based differences.

The purpose of this analysis is to create a professional financial research project suitable for a GitHub portfolio and Upwork portfolio presentation.

### What is the DAX?

The DAX (Deutscher Aktienindex) is Germany's leading stock market index. It tracks the performance of the largest publicly traded companies listed on the Frankfurt Stock Exchange.

The index serves as a benchmark for the German economy and provides investors with exposure to major sectors including technology, industrials, financial services, telecommunications, healthcare, and consumer goods.


## 2. Data Collection

Historical stock price data will be collected from Yahoo Finance using the yfinance library.

The analysis focuses on five major German companies:

- SAP (Technology)
- Siemens (Industrials)
- Allianz (Financial Services)
- Deutsche Telekom (Telecommunications)
- Mercedes-Benz (Automotive)

A minimum of five years of daily historical data will be used for analysis.
